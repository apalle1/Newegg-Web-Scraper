```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
# open up connection and download the webpage
my_url = 'https://www.newegg.com/Video-Cards-Video-Devices/Category/ID-38?Tpk=graphics%20card'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()

# html parsing
page_soup = soup(page_html, "html.parser")

# To find the header of the page
print(page_soup.h1)
```

```python
# We first inspect any element on the page and then find the body tag. 
print(page_soup.body)
```

* Notice that this body tag can go even further into an A tag or SPAN tag. 


```python
print(page_soup.body.a)
print(page_soup.body.span)
```

* Next we select any graphics card on the page and try to inspect that element.
* I then wish to grab the entire container of the selected graphics card.
* Bcoz I know that graphics card container contains info such as the original price, sale price, manufacturer, review, card image.
* We need to find what it is that is containing all of the information related to a graphics card.
* We notice that the "<div class="item-container" contains and houses all of the items inside. 
* So basically I need to set a loop. 
* I would write my script first on how to parse one graphics card, and then once I'm done with that, I can loop through all of the "<div class="item-container" to parse every single graphics card into my data file.
* I go to my page_soup, there is a function called findAll. Now what do I want to find ?
I want to find all div's that have the class "item-container".
To findAll, I would say, find me all div's and I also have feed it an object. 
An object says what is the name of the tag that you're looking for ? So it's a class. If it was an ID, I would put Id.


```python
# grabs each product
containers = page_soup.findAll("div", {"class":"item-container"})
print(len(containers))
print('---------------------------------------------------------------------------------------------------------------')
print(containers[0])
```

* I'm gonna throw containers[0] into container. Later, I'm going to do a for loop for every container in containers.
So I want to make sure it works once before I even build the loop.

* Notice that Title is inside of square bracket which means it's inside of a list. Title[0] grabs the first object in the list.
Inside the first object, I want to grab the text.

* Notice that the output obtained from shipcost[0].text. It looks like there is a lot of open space.
So in this case I want to clean it up a little bit bcoz I just want the text. So in this case I will use strip().
This removes blank spaces before and after. 

```python
container = containers[0]

brand = container.find("div", {"class":"item-info"}).div.a.img['title']
print(brand)
print('')

title = container.findAll("a", {"class":"item-title"})
print(title)
print('')
print(title[0])
print('')
print(title[0].text)

shipcost = container.findAll("li", {"class":"price-ship"})
print(shipcost)
print('')
print(shipcost[0])
print('')
print(shipcost[0].text)
print('')
print(shipcost[0].text.strip())
```

```python
for container in containers:
    brand = container.find("div", {"class":"item-info"}).div.a.img['title']
    
    title = container.findAll("a", {"class":"item-title"})
    product_name = title[0].text
    
    shipcost = container.findAll("li", {"class":"price-ship"})
    shipping = shipcost[0].text.strip()
    
    print("brand: " + brand)
    print("product_name: " + product_name)
    print("shipping: " + shipping)
```

```python
filename = "newegg_product.csv"
headers = "brand, product_name, shipping \n"

f = open(filename, 'w')
f.write(headers)
```

Let me double check something really quick. See if my strings are clean. And no it is not.
Notice that product_name has commas inside of them. So what that's gonna do is create extra columns inside of my CSV file.



```python
for container in containers:
    brand = container.find("div", {"class":"item-info"}).div.a.img['title']
    
    title = container.findAll("a", {"class":"item-title"})
    product_name = title[0].text
    
    shipcost = container.findAll("li", {"class":"price-ship"})
    shipping = shipcost[0].text.strip()
    
    print("brand: " + brand)
    print("product_name: " + product_name)
    print("shipping: " + shipping)

    f.write(brand + "," + product_name.replace(",", "|") + "," + shipping + "\n")
    
f.close()
```
