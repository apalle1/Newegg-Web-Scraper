```python
import numpy as np
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
# open up connection and download the webpages
url_array = ['https://www.newegg.com/Video-Cards-Video-Devices/Category/ID-38?Tpk=graphics%20card', 
'https://www.newegg.com/Product/ProductList.aspx?Submit=ENE&DEPA=0&Order=BESTMATCH&Description=cardio+equipment&N=-1&isNodeId=1']

page_html = []
for url in url_array:
    uClient = uReq(url)
    page_html.append(uClient.read())
    uClient.close()
```

```python
items_array = []
for key, element in enumerate(page_html):
    page_soup = soup(element, "html.parser")
    containers = page_soup.findAll("div", {"class":"item-container"})
    item_array = []
    
    for container in containers:
        brand = container.find("div", {"class":"item-info"}).div.a.img['title']
    
        title = container.findAll("a", {"class":"item-title"})
        product_name = title[0].text
    
        shipcost = container.findAll("li", {"class":"price-ship"})
        shipping = shipcost[0].text.strip()
    
        print("brand: " + brand)
        print("product_name: " + product_name)
        print("shipping: " + shipping)

        item_array.append(brand + "," + product_name.replace(",", "|") + "," + shipping + "\n") 
    
    items_array.append(item_array)
```

```python
str1 = ''.join(items_array[0])
print(str1)
```

```python
fileName = ['graphic_card.csv', 'cardio_equipment.csv']
headers = "brand, product_name, shipping \n"

for index in range(len(fileName)):
    csv_file = open(fileName[index], 'w')
    csv_file.write(headers)
    csv_file.write(''.join(items_array[index]))
    csv_file.close()
```
